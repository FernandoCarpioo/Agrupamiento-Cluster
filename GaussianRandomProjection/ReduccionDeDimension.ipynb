```python
from sklearn.datasets import load_digits
from sklearn.random_projection import GaussianRandomProjection
import matplotlib.pyplot as plt

digits = load_digits()
X = digits.data
y = digits.target

plt.figure(figsize=(8, 3))
for i in range(10):
    plt.subplot(2, 5, i + 1)
    plt.imshow(digits.images[i], cmap='gray')
    plt.title(f"Dgito: {digits.target[i]}")
    plt.axis('off')

plt.suptitle("Muestras del dataset original")
plt.tight_layout()
plt.show()

plt.figure(figsize=(9,4))

plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='tab10', s=40, vmin=0, vmax=9)
plt.xlabel("Intensidad pxel 1")
plt.ylabel("Intensidad pxel 2")
plt.title("Digits dataset - Antes de reduccin")

print("Forma del dataset:", X.shape)

GRP = GaussianRandomProjection(n_components=2)
X_red = GRP.fit_transform(X)

plt.subplot(1, 2, 2)
scatter = plt.scatter(X_red[:,0], X_red[:,1], c=y, cmap='tab10', s=30, vmin=0, vmax=9)
plt.xlabel("Componente Principal 1")
plt.ylabel("Componente Principal 2")
plt.title("GaussianRandomProjection - Digits dataset")
plt.colorbar(scatter, label="Dgito")

plt.tight_layout()
plt.show()
```

```python
from sklearn.datasets import load_wine
from sklearn.random_projection import GaussianRandomProjection
import matplotlib.pyplot as plt

wine = load_wine()
X = wine.data
y = wine.target
feature_names = wine.feature_names

plt.figure(figsize=(9,4))

plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 9], c=y, cmap='rainbow', s=50)
plt.xlabel(feature_names[0])
plt.ylabel(feature_names[9])
plt.title("Wine dataset - Antes de reduccin")

print("Forma del dataset:", X.shape)

GRP = GaussianRandomProjection(n_components=2, random_state=9)
X_red = GRP.fit_transform(X)

plt.subplot(1, 2, 2)
scatter = plt.scatter(X_red[:,0], X_red[:,1], c=y, cmap='rainbow', s=50)
plt.xlabel("Componente Principal 1")
plt.ylabel("Componente Principal 2")
plt.title("GaussianRandomProjection - Wine dataset")

plt.tight_layout()
plt.show()
```
